## 1. DATA ACQUISITION
Let's download the dataset using Kaggle API:

```python
import os
import shutil

print("Organizing dataset into project structure...")

# Source folders (inside dogcat folder)
source_train = "dogcat/train"
source_test = "dogcat/test"
source_validation = "dogcat/validation"

# Destination folders (our data directory)
dest_train = "data/train"
dest_test = "data/test"
dest_validation = "data/validation"

# Move train folder
if os.path.exists(source_train):
    if os.path.exists(dest_train):
        shutil.rmtree(dest_train)  # Remove old if exists
    shutil.copytree(source_train, dest_train)
    print("Copied train/ → data/train/")

# Move test folder
if os.path.exists(source_test):
    if os.path.exists(dest_test):
        shutil.rmtree(dest_test)
    shutil.copytree(source_test, dest_test)
    print("Copied test/ → data/test/")

# Move validation folder
if os.path.exists(source_validation):
    if os.path.exists(dest_validation):
        shutil.rmtree(dest_validation)
    shutil.copytree(source_validation, dest_validation)
    print("Copied validation/ → data/validation/")
```
